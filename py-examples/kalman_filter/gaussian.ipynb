# Components

```python
rDm = np.array([[2, -1]]).T
a = np.pi/4.0
rRm =  np.array([[np.cos(a), -np.sin(a)], [np.sin(a), np.cos(a)]])
# Display the matrix using LaTeX formatting
sp.init_printing()  # Enables pretty printing
display(Matrix(rRm))
display(Matrix(rDm))
```

# Transformation Matrix

```python
rTm = np.matrix(np.block([[rRm,rDm],[0,0,1]]))

display(Matrix(rTm))
display(Matrix(rTm))
```
